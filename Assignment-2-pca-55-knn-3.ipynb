```python
# import standard libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# load train data
df = pd.read_csv('../input/classification-with-non-deep-classifiers/train.csv')
df.head()
```

```python
df.shape
```

```python
#drop the label column from the training data
X = df.drop(['label'],axis=1)
```

```python
#store the label data
y = df.label
```

```python
from sklearn.model_selection import train_test_split

#split dataset into training data (70%) and test data (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
# Dimensionality reduction - PCA
from sklearn.decomposition import PCA

pca = PCA(55)
X_train_pca = pca.fit_transform(X_train)
X_train_pca.shape
```

```python
# kNN classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_pca,y_train)
```

```python
#PCA on test data
X_test_pca = pca.transform(X_test)
X_test_pca.shape
```

```python
#Test model
knn.score(X_test_pca, y_test)
```

```python
# Prepare the submission
test_data = pd.read_csv('../input/classification-with-non-deep-classifiers/test.csv')
test_data_numpy = test_data.to_numpy()
test_data_pca = pca.transform(test_data)
predicted_labels = knn.predict(test_data_pca)
id = [i for i in range(test_data_numpy.shape[0])]
test_submission = pd.DataFrame({'id': id, 'label': predicted_labels})

test_submission.to_csv('submission.csv', index=False)
```
